```python
detail_stations = pd.DataFrame({
    'Station':['a' for i in range(1977)],
    'Catchment_No': [i for i in range(1977)],
    'District': ['a' for i in range(1977)],
    'Latitude': ['a' for i in range(1977)],
    'Longitude': ['a' for i in range(1977)]})
```

```python
i=0
station=pre_station=''
null_v=[False,False,False,False,False]
null_one=False
index=0
while i<df.shape[0]:
    print(i,df.shape,detail_stations.shape)
    station=df.iloc[i,0]
    if station==pre_station:
        if not null_one:
            pass
#             df.drop(df.index[i])
        else:
            if null_v[1]:
                if pd.notna(df.iloc[i,1]):
                    df.iloc[index,1]=df.iloc[i,1]
            if null_v[2]:
                if pd.notna(df.iloc[i,2]):
                    df.iloc[index,2]=df.iloc[i,2]
            if null_v[3]:
                if pd.notna(df.iloc[i,3]):
                    df.iloc[index,3]=df.iloc[i,3]
            if null_v[4]:
                if pd.notna(df.iloc[i,4]):
                    df.iloc[index,4]=df.iloc[i,4]
    else:
        detail_stations.iloc[index]=df.iloc[i]
        index+=1
        j=0
        while j<=4:
            if pd.isnull(df.iloc[i,j]):
                null_v[j]=True
            j+=1
        null_one=False
        for k in null_v:
            if k:
                null_one=True
                break
        pre_station=station
    i+=1
```
